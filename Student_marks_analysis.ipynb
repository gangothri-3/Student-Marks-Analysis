```python
import pandas as pd

# Read uploaded file
df = pd.read_csv("student_data.csv")  # Adjust if filename differs
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Correlation matrix
correlation = df[['MathScore', 'ReadingScore', 'WritingScore']].corr()

# Heatmap
plt.figure(figsize=(6,4))
sns.heatmap(correlation, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Between Scores")
plt.show()

# Pairplot
sns.pairplot(df, hue="Result", vars=["MathScore", "ReadingScore", "WritingScore"])
plt.suptitle("Pairwise Score Distributions", y=1.02)
plt.show()

# Boxplots
plt.figure(figsize=(12, 4))
for i, col in enumerate(["MathScore", "ReadingScore", "WritingScore"]):
    plt.subplot(1, 3, i+1)
    sns.boxplot(x="Result", y=col, data=df)
    plt.title(f"{col} by Result")
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

# Features & Label
X = df[["MathScore", "ReadingScore", "WritingScore"]]
y = df["Result"]

# Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

# Train KNN
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

# Accuracy
print("Model Accuracy:", knn.score(X_test, y_test))
```

```python
def predict_new_student(math, reading, writing):
    data = [[math, reading, writing]]
    scaled_data = scaler.transform(data)
    prediction = knn.predict(scaled_data)[0]
    return prediction

# Example: Take input from user
math = float(input("Enter Math Score: "))
reading = float(input("Enter Reading Score: "))
writing = float(input("Enter Writing Score: "))

result = predict_new_student(math, reading, writing)
print("Predicted Result:", result)
```
